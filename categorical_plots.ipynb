#Stripplot and swarnplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they donâ€™t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).




Imagine you are a school teacher checking math marks of students from three sections: A, B, and C.
A strip plot is like you pin all studentsâ€™ marks on a board under their section name. If many students scored similar marks, the pins may overlap and look messy, but you can still see every single studentâ€™s score. Itâ€™s simple and quick, and when you add it with a box plot, you can see both the overall result and individual students together.

Now think of a swarm plot as a smart teacher who notices the pins are overlapping and slightly moves them sideways so every pin is visible. No marks are hidden, and you can clearly see where most students scored and how the marks are spread. But if there are too many students, arranging all those pins neatly takes time and the board starts to look crowded. So swarm plots are great for small classes, but not ideal for very large groups.

```python
sns.stripplot(x='day',y='total_bill',data=df) # means days ke strip bni usi me jitne
# jis din me jo values mili unhe us strip ke andar plot kr diya
```

```python
sns.swarmplot(x='day',y='total_bill',data=df) # ye sare points ko jha data spread hoga wha pr distinct krke dikhayega
```

```python
#if we want to create violinplot and swarnplot on same canvas graph
sns.violinplot(x='tip',y='day',data=df)
sns.swarmplot(x='tip',y='day',data=df)
```
