```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_id',how='inner')# inner emp id ke common values uske bases pr merge # just like join in sql and common column is emp_id as we are merging the common enteries using a common column
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')#aab bs emp id ke basis pr merge krenge unke common pr nhi
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')#left join
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')# right join
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
```

```python
df
```

```python
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='median')# index- row ka index column ->column index of products# jiska jiske sath connection hoga ie east ka a ke sath to mean nikl kr ayega else nan
```

```python
pivot2=pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
```

```python
pivot2#works seperately for 2 input values and use numeric values(must) in values
```

```python
#cross Tabs
```

```python
pd.crosstab(df['Region'],df['Product'])#use for counting functions and pivot is used for aggregation functions
```
