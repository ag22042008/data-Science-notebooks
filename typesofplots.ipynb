```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns
#distribution plots are created for quantitative variables(uncategorised data)-age salary
 #histogram
 #join plot
 #pairplot
 #rugplot
```

```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
df['size'].unique()# categoricAL DATA  size is also a qualitative data
```

```python
sns.displot(df['tip'],bins=20)# width indicates bin size
# height indicates no of values in that bin
```

```python

sns.histplot(df['total_bill'],bins=30,kde=True)#kernel density estimator these tell us that which direction data is correctly skewed left skewed right or normal
```

```python
plt.subplot(1,2,1)
sns.histplot(df['tip'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['total_bill'],bins=30,kde=True)
```

```python
# joint plot if we want to compare two data we use joint plot
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter',color='green')
```

```python
# different kinds
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde',color='green')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex',color='violet')#jha data jyada wha dense color
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg',color='red') # kind of mimics linear regression model
```

```python
#pair plots
sns.pairplot(df,hue='size',palette='rainbow')# hue if we want to check which category
#has more tips eg if female has given more tips or male similarly (or categorical data division)
#for total bills also
# numerical values ka comparison hoga all numerical values ka
```

```python
#rug plot
sns.rugplot(df['tip'])
```
