```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten
```

```python
(X_train,Y_train),(X_test,Y_test)=keras.datasets.mnist.load_data()
```

```python
#we need to make all the values in a single range btw 0 and 1 as they are bound
X_train=X_train/255#(Normalisation)
X_test=X_test/255
```

# we need to make the array flatten to make all the rows side by side

```python
model=Sequential()
model.add(Flatten(input_shape=(28,28)))#we need to first specify the input shape of the data given
model.add(Dense(32,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
history=model.fit(X_train,Y_train,epochs=25,validation_split=0.2)
```

```python
y_probaility=model.predict(X_test)# it will run on every test image and display what is the probability of 0 1 2 3 4 5 6 and so on to occur as an image
```

```python
# but we need to know for every array konse index position pr value max hai
y_pred=y_probaility.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])#overfitting there it is
```
