# object oriented way
# this a way to represent  or plot an axis in fully canvas or control the canvas fully in our way
Object Oriented way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).


```python
fig=plt.figure(figsize=[6,8],dpi=200)#[6,8]->600*800(width*height) for we can perform changes in fig size and dpi fig obj is created
fig.set_facecolor('lightgray')
axis1=fig.add_axes([0.1,0.5,0.9,0.7])# 1->left se distance 2->bottom se distance 3-> width 4-> height
axis1.plot(x,x**2,color='lightblue')# axis pr hi plot kr skte h

axis2=fig.add_axes([0.2,0.7,0.6,0.4])
axis2.plot(x,x**3)
```
