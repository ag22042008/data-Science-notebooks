```python
# use lmplot to create regression plots
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

x-> axis the basis on which we are predicting something
y->the prediction on basis of x


```python
# implot
tips=sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips,scatter_kws={'color': 'lightgreen'},line_kws={'color':'red'})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',scatter_kws={'color': 'lightgreen'})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})#default value 50 of s
```
